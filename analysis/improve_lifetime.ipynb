```julia
include("../src/utilities.jl");
```

```julia
using PyPlot
```

```julia
data_dir = "../data/improve_lifetime/";
```

```julia
t1, state1, params1 = load_state_jld("$(data_dir)state_1_no_drive_no_sz.jld2")
t2, state2, params2 = load_state_jld("$(data_dir)state_2_no_drive_sz_up.jld2")
t3, state3, params3 = load_state_jld("$(data_dir)state_3_no_drive_sz_down.jld2");
t4, state4, params4 = load_state_jld("$(data_dir)state_4_no_drive_sz_up_repeated.jld2");
```

```julia
purity1 = state_purity(state1)
purity2 = state_purity(state2)
purity3 = state_purity(state3)
purity4 = state_purity(state4);
```

```julia
t1_s = t1.*params1[:sys_params]["g_2"]
t2_s = t2.*params2[:sys_params]["g_2"]
t3_s = t3.*params3[:sys_params]["g_2"]
t4_s = t4.*params4[:sys_params]["g_2"];
```

```julia
PyPlot.plot(t1_s, purity1, label=L"\sigma_x = 1", color="black")
PyPlot.plot(t2_s, purity2, label=L"\sigma_z = 1", color="red")
PyPlot.plot(t3_s, purity3, label=L"\sigma_z = -1", color="blue")
PyPlot.plot(t3_s, purity3, label=L"\sigma_z = 1 \text{ repeated} ", color="red", linestyle = "dashed")
PyPlot.ylim(0.68, 1.01)
PyPlot.xlim(-0.05, 1.0)
PyPlot.xlabel(L"g t", fontsize=15)
PyPlot.ylabel("Purity", fontsize=15)
PyPlot.plot(grid=false, framestyle=:box)
PyPlot.xticks(ticks=[0.0, 0.5, 1.0], fontsize=11)
PyPlot.yticks(ticks=[0.7, 0.8, 0.9, 1.0], fontsize=11)
PyPlot.legend(loc="best", fontsize=11)
```
